```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

class PatientDataAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.file_path)
        #print(" Data Loaded Successfully.")
        
    def basic_info(self):
        print("\n--- Basic Info ---")
        print(self.data.info())
        print("\n--- Description ---")
        print(self.data.describe(include='all'))
        print("-----------------------------------------------------------------------------")

    def age_distribution(self):
        plt.figure(figsize=(6, 4))
        sns.histplot(self.data['age'], bins=15, kde=True, color='skyblue')
        plt.title('Age Distribution')
        plt.tight_layout()
        plt.show()
        print("-----------------------------------------------------------------------------")

    def bmi_distribution(self):
        plt.figure(figsize=(6, 4))
        sns.histplot(self.data['bmi'], bins=15, kde=True, color='coral')
        plt.title('BMI Distribution')
        plt.tight_layout()
        plt.show()
        print("-----------------------------------------------------------------------------")

    def lifestyle_analysis(self):
        plt.figure(figsize=(10, 4))
        plt.subplot(1, 2, 1)
        sns.countplot(x='smoking', data=self.data, palette='coolwarm')
        plt.title('Smoking Status')
        print("-----------------------------------------------------------------------------")

        plt.subplot(1, 2, 2)
        sns.countplot(x='alcohol', data=self.data, palette='coolwarm')
        plt.title('Alcohol Consumption')
        plt.tight_layout()
        plt.show()
        print("-----------------------------------------------------------------------------")

    def exercise_analysis(self):
        plt.figure(figsize=(6, 4))
        sns.histplot(self.data['exercise_hours_per_week'], bins=10, kde=True, color='green')
        plt.title('Exercise Hours Per Week')
        plt.tight_layout()
        plt.show()
        print("-----------------------------------------------------------------------------")

    def stress_by_gender(self):
        plt.figure(figsize=(6, 4))
        sns.countplot(x='stress_level', hue='gender', data=self.data, palette='Set1')
        plt.title('Stress Level by Gender')
        plt.tight_layout()
        plt.show()

    def correlation_heatmap(self):
        numerical = self.data.select_dtypes(include='number')
        plt.figure(figsize=(6, 4))
        sns.heatmap(numerical.corr(), annot=True, cmap='coolwarm')
        plt.title("Correlation Heatmap")
        plt.tight_layout()
        plt.show()

# ---- USAGE ----
analyzer = PatientDataAnalyzer("updated_patients.csv")
analyzer.load_data()
analyzer.basic_info()
analyzer.age_distribution()
analyzer.bmi_distribution()
analyzer.lifestyle_analysis()
analyzer.exercise_analysis()
analyzer.stress_by_gender()
analyzer.correlation_heatmap()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

class PatientDataAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.file_path)

    def gender_visit_count(self):
        print("\n Gender-wise Patient Count:")
        print(self.data['gender'].value_counts())

    def alcohol_smoking_status(self):
        print("\n Patients who Consume Alcohol:")
        print(self.data[self.data['alcohol'] == 'Yes'][['patient_id', 'gender']])
        
        print("\n Patients who Smoke:")
        print(self.data[self.data['smoking'] == 'Yes'][['patient_id', 'gender']])

    def family_history_heart_risk(self):
        total = len(self.data)
        with_family_history = self.data[self.data['family_history'] == 'Yes']
        risk_percent = (len(with_family_history) / total) * 100
        print(f"\n Family History & Heart Attack Risk:")
        print(f"Total Patients: {total}")
        print(f"With Family History: {len(with_family_history)}")
        print(f"Chances of Heart Attack (with family history): {risk_percent:.2f}%")

    def heart_risk_by_gender(self):
        risk_by_gender = self.data[self.data['family_history'] == 'Yes'].groupby('gender').size()
        print("\nGender-wise Heart Attack Risk (Based on Family History):")
        print(risk_by_gender)
        self._beautiful_heart_risk_chart(risk_by_gender)

    def _beautiful_heart_risk_chart(self, risk_by_gender):
        plt.figure(figsize=(6, 4))
        sns.set_style("whitegrid")  

        # Convert to DataFrame
        data = risk_by_gender.reset_index()
        data.columns = ['gender', 'count']

        # Thin bars with soft but vivid color
        ax = sns.barplot(
            x='gender',
            y='count',
            data=data,
            palette=['#3498DB', '#E74C3C'],  
            width=0.2  
        )

        # Value labels
        for p in ax.patches:
            ax.annotate(f'{int(p.get_height())}', 
                        (p.get_x() + p.get_width() / 2., p.get_height() + 0.2), 
                        ha='center', va='bottom', fontsize=12, fontweight='bold', color='black')

        plt.title(" Heart Risk by Gender (Family History)", fontsize=14, fontweight='bold')
        plt.xlabel("Gender", fontsize=12)
        plt.ylabel("Number of Patients", fontsize=12)
        plt.tight_layout()
        plt.show()

# -------------------- USAGE --------------------

analyzer = PatientDataAnalyzer("updated_patients.csv")
analyzer.load_data()
analyzer.gender_visit_count()
analyzer.alcohol_smoking_status()
analyzer.family_history_heart_risk()
analyzer.heart_risk_by_gender()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("updated_patients.csv")

# Filter patients with family history (heart risk group)
heart_risk = df[df['family_history'] == 'Yes']

# Group by gender
risk_by_gender = heart_risk['gender'].value_counts()

# Create a pie chart
plt.figure(figsize=(5, 5))
colors = ['Blue', 'red']  # Blue for Male, Red for Female
plt.pie(
    risk_by_gender,
    labels=risk_by_gender.index,
    autopct='%1.1f%%',
    startangle=90,
    colors=colors,
    textprops={'fontsize': 12}
)
plt.title("Heart Risk by Gender (Family History)", fontsize=14, fontweight='bold')
plt.axis('equal')  # Make it a circle
plt.tight_layout()
plt.show()
```

```python
class PatientVisualizer:
    def __init__(self, data):
        self.data = data

    def bar_chart_gender_distribution(self):
        plt.figure(figsize=(6, 4))
        ax = sns.countplot(
            x='gender', 
            data=self.data, 
            palette='Set2'
        )

        # Make bars thinner by reducing bar width
        for bar in ax.patches:
            bar.set_width(0.4)  # Default is 0.8  reduce to make thinner
            bar.set_edgecolor('black')  # Optional: add black border for clarity

        # Add value labels
        for p in ax.patches:
            ax.annotate(
                f'{p.get_height()}',
                (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center',
                fontsize=12, color='black',
                xytext=(0, 8),
                textcoords='offset points'
            )

        # Styling
        ax.set_title('Gender Distribution', fontsize=14, weight='bold')
        ax.set_xlabel('Gender', fontsize=12)
        ax.set_ylabel('Count', fontsize=12)
        sns.despine()
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.tight_layout()
        plt.show()

# Create an instance and generate the thin bar chart
visualizer = PatientVisualizer(df)
visualizer.bar_chart_gender_distribution()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import io

class PatientDataAnalyzer:
    def __init__(self, file_path):
        self.file_path = file_path
        self.data = None

    def load_data(self):
        self.data = pd.read_csv(self.file_path)

    def basic_info(self, pdf):
        # Capture .info() output
        buffer = io.StringIO()
        self.data.info(buf=buffer)
        info_str = buffer.getvalue()
        desc_str = str(self.data.describe(include='all'))

        # Combine into one block of text
        combined_text = f"--- Basic Info ---\n{info_str}\n\n--- Description ---\n{desc_str}"

        # Plot text
        plt.figure(figsize=(8.5, 11))
        plt.axis('off')
        plt.title("Basic Dataset Information", fontsize=16)
        plt.text(0.01, 0.95, combined_text, fontsize=7, va='top', family='monospace')
        pdf.savefig()
        plt.close()

    def age_distribution(self, pdf):
        if 'age' in self.data.columns:
            plt.figure(figsize=(6, 4))
            sns.histplot(self.data['age'].dropna(), bins=15, kde=True, color='skyblue')
            plt.title('Age Distribution')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

    def bmi_distribution(self, pdf):
        if 'bmi' in self.data.columns:
            plt.figure(figsize=(6, 4))
            sns.histplot(self.data['bmi'].dropna(), bins=15, kde=True, color='lightgreen')
            plt.title('BMI Distribution')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

    def gender_distribution(self, pdf):
        if 'gender' in self.data.columns:
            plt.figure(figsize=(6, 4))
            sns.countplot(x='gender', data=self.data, palette='pastel')
            plt.title('Gender Distribution')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

    def correlation_heatmap(self, pdf):
        try:
            corr = self.data.corr(numeric_only=True)
            if not corr.empty:
                plt.figure(figsize=(6, 4))
                sns.heatmap(corr, annot=True, cmap='coolwarm')
                plt.title('Correlation Heatmap')
                plt.tight_layout()
                pdf.savefig()
                plt.close()
        except Exception as e:
            print("Correlation heatmap skipped:", e)

    def condition_count(self, pdf):
        if 'condition' in self.data.columns:
            plt.figure(figsize=(6, 4))
            sns.countplot(y='condition', data=self.data, order=self.data['condition'].value_counts().index, palette='Set2')
            plt.title('Condition Count')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

    def avg_bmi_by_condition(self, pdf):
        if 'condition' in self.data.columns and 'bmi' in self.data.columns:
            avg_bmi = self.data.groupby('condition')['bmi'].mean().sort_values()
            plt.figure(figsize=(6, 4))
            avg_bmi.plot(kind='barh', color='coral')
            plt.title('Average BMI by Condition')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

    def age_bmi_scatter(self, pdf):
        if 'age' in self.data.columns and 'bmi' in self.data.columns:
            plt.figure(figsize=(6, 4))
            sns.scatterplot(data=self.data, x='age', y='bmi', hue='gender')
            plt.title('Age vs BMI Scatter Plot')
            plt.tight_layout()
            pdf.savefig()
            plt.close()

def generate_patient_report():
    analyzer = PatientDataAnalyzer("updated_patients.csv")  # Make sure this file exists
    analyzer.load_data()

    with PdfPages("Patients_Report.pdf") as pdf:
        analyzer.basic_info(pdf)
        analyzer.age_distribution(pdf)
        analyzer.bmi_distribution(pdf)
        analyzer.gender_distribution(pdf)
        analyzer.correlation_heatmap(pdf)
        analyzer.condition_count(pdf)
        analyzer.avg_bmi_by_condition(pdf)
        analyzer.age_bmi_scatter(pdf)

    print(" Patients_Report.pdf created successfully!")

# Run the report
generate_patient_report()
```
